```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='green',linestyle='--',marker='o')
plt.title('My Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = [1,2,3]
y1 = [1,5,9]
y2 = [2,6,-5]
plt.plot(x,y1,label='Squares')
plt.plot(x,y2,label='Linear')
plt.legend()
plt.show()
```

```python
categories = ['A','B','C','D','E']
values = [5,7,3,10,4]
plt.bar(categories,values)
plt.show()
```

```python
data = [1,1,2,3,3,3,4,4,5]
plt.hist(data,bins=10)
plt.show()
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.scatter(x,y)
plt.show()
```
